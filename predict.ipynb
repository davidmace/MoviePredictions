```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# load data
movie_data = pd.io.excel.read_excel( open('moviedata_all.xlsx','r') )
```

```python
# shuffle data
movie_data = movie_data.reindex(np.random.permutation(movie_data.index))
```

```python
# split into train, validate, test data
train_data = movie_data.ix[:3000]
validate_data = movie_data.ix[3000:3400]
test_data = movie_data.ix[3400:]
```

```python
# baseline predictor... gross_international=f(budget) where f=ax+0 ... this is what they currently have
xtrain = np.array(train_data['budget'])
ytrain = np.array(train_data['total gross'])
coeff = np.transpose([xtrain])
((m), _, _, _) = np.linalg.lstsq(coeff, ytrain)
print m
```

```python
# error calculation
xtest = np.array(test_data['budget'])
ytest = np.array(test_data['total gross'])
yhat=m*xtest
err = np.sqrt( np.sum( (yhat-ytest)**2 ) )/ np.shape(xtest)[0]
print err
fig, axes = plt.subplots(figsize=(20,20), dpi=100)
plt.scatter(xtest, ytest)

plt.plot(xtest,yhat,color='red')
plt.show()
```

```python
# get a reliable error score

toterr = 0
for i in range(1000) :
    
    # shuffle data
    movie_data = movie_data.reindex(np.random.permutation(movie_data.index))

    # split into train, validate, test data
    train_data = movie_data.ix[:3000]
    validate_data = movie_data.ix[3000:3400]
    test_data = movie_data.ix[3400:]

    # baseline predictor... gross_international=f(budget) where f=ax+0 ... this is what they currently have
    xtrain = np.array(train_data['budget'])
    ytrain = np.array(train_data['total gross'])
    coeff = np.transpose([xtrain])
    ((m), _, _, _) = np.linalg.lstsq(coeff, ytrain)

    # error calculation
    xtest = np.array(test_data['budget'])
    ytest = np.array(test_data['total gross'])
    yhat=m*xtest
    err = np.sqrt( np.sum( (yhat-ytest)**2 ) )/ np.shape(xtest)[0]
    toterr+=err
    
print toterr/1000
```
