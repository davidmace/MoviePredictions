```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# load data
movie_data = pd.io.excel.read_excel( open('moviedata_all.xlsx','r') )
```

```python
# shuffle data
movie_data = movie_data.reindex(np.random.permutation(movie_data.index))
```

```python
# split into train, validate, test data
train_data = movie_data.ix[:3000]
validate_data = movie_data.ix[3000:3400]
test_data = movie_data.ix[3400:]
```

```python
train_inds = np.where( np.array(train_data['budget'])*np.array(train_data['inflation correction'])>50000000 )[0]
test_inds = np.where( np.array(test_data['budget'])*np.array(test_data['inflation correction'])>50000000 )[0]
```

```python
# with dow corrected for inflation

trainerr = 0
testerr = 0
for i in range(10000) :
    
    # shuffle data
    movie_data = movie_data.reindex(np.random.permutation(movie_data.index))

    # split into train, validate, test data
    train_data = movie_data.ix[:3000]
    validate_data = movie_data.ix[3000:3400]
    test_data = movie_data.ix[3400:]

    # train
    xtrain1 = np.array(train_data['budget'])*np.array(train_data['inflation correction'])
    null_ind = np.where( np.array( pd.isnull( train_data['budget'] ) ) )
    xtrain1[null_ind]=np.nanmean(xtrain1)

    xtrain2 = np.array(train_data['dow'])*np.array(train_data['inflation correction'])
    null_ind = np.where( np.array( pd.isnull( train_data['dow'] ) ) )
    xtrain2[null_ind]=np.nanmean(xtrain2)

    ytrain = np.array(train_data['total gross'])*np.array(train_data['inflation correction'])
    coeff = np.transpose([xtrain1,xtrain2])
    ((a,b), _, _, _) = np.linalg.lstsq(coeff, ytrain)
    
    # test
    xtest1 = np.array(test_data['budget'])*np.array(test_data['inflation correction'])
    null_ind = np.where( np.array( pd.isnull( test_data['budget'] ) ) )
    xtest1[null_ind]=np.nanmean(xtest1)

    xtest2 = np.array(test_data['dow'])*np.array(test_data['inflation correction'])
    null_ind = np.where( np.array( pd.isnull( test_data['dow'] ) ) )
    xtest2[null_ind]=np.nanmean(xtest2)

    ytest = np.array(test_data['total gross'])*np.array(test_data['inflation correction'])

    yhat=a*xtrain1+b*xtrain2
    err = np.sqrt( np.sum( ((yhat-ytrain)/ytrain)**2 ) )/ np.shape(xtrain1)[0]
    #print err
    trainerr+=err
    
    # error calculation
    yhat=a*xtest1+b*xtest2
    err = np.sqrt( np.sum( ((yhat-ytest)/ytest)**2 ) )/ np.shape(xtest1)[0]
    #print err
    testerr+=err
    
print trainerr/10000, testerr/10000
```

```python
# get a reliable error score

trainerr = 0
testerr = 0
for i in range(10000) :
    
    # shuffle data
    movie_data = movie_data.reindex(np.random.permutation(movie_data.index))

    # split into train, validate, test data
    train_data = movie_data.ix[:3000]
    validate_data = movie_data.ix[3000:3400]
    test_data = movie_data.ix[3400:]

    # train
    xtrain1 = np.array(train_data['budget'])*np.array(train_data['inflation correction'])
    null_ind = np.where( np.array( pd.isnull( train_data['budget'] ) ) )
    xtrain1[null_ind]=np.nanmean(xtrain1)

    xtrain2 = np.array(train_data['dow'])
    null_ind = np.where( np.array( pd.isnull( train_data['dow'] ) ) )
    xtrain2[null_ind]=np.nanmean(xtrain2)

    ytrain = np.array(train_data['total gross'])*np.array(train_data['inflation correction'])
    coeff = np.transpose([xtrain1,xtrain2,xtrain1*xtrain2])
    ((a,b,c), _, _, _) = np.linalg.lstsq(coeff, ytrain)
    
    # test
    xtest1 = np.array(test_data['budget'])*np.array(test_data['inflation correction'])
    null_ind = np.where( np.array( pd.isnull( test_data['budget'] ) ) )
    xtest1[null_ind]=np.nanmean(xtest1)

    xtest2 = np.array(test_data['dow'])
    null_ind = np.where( np.array( pd.isnull( test_data['dow'] ) ) )
    xtest2[null_ind]=np.nanmean(xtest2)

    ytest = np.array(test_data['total gross'])*np.array(test_data['inflation correction'])

    yhat=a*xtrain1+b*xtrain2+c*xtrain1*xtrain2
    err = np.sqrt( np.sum( ((yhat-ytrain)/ytrain)**2 ) )/ np.shape(xtrain1)[0]
    #print err
    trainerr+=err
    
    # error calculation
    yhat=a*xtest1+b*xtest2+c*xtest1*xtest2
    err = np.sqrt( np.sum( ((yhat-ytest)/ytest)**2 ) )/ np.shape(xtest1)[0]
    #print err
    testerr+=err
    
print trainerr/10000, testerr/10000
```
